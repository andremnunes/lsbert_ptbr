## **BERTimbau**

```python
def baixarBERTimbau():
  from transformers import AutoTokenizer  # Or BertTokenizer
  from transformers import AutoModelForPreTraining  # Or BertForPreTraining for loading pretraining heads
  from transformers import AutoModel  # or BertModel, for BERT without pretraining heads

  model = AutoModelForPreTraining.from_pretrained('neuralmind/bert-base-portuguese-cased')
  tokenizer = AutoTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased', do_lower_case=True)

  output_model_file = "/content/drive/MyDrive/tcc/BERT-LS-master/bert/bert-base-portuguese-uncased-pytorch_model.bin"
  output_config_file = "/content/drive/MyDrive/tcc/BERT-LS-master/bert/bert-base-portuguese-uncased-config.json"
  output_vocab_dir = "/content/drive/MyDrive/tcc/BERT-LS-master/bert/"
  output_vocab_file = "/content/drive/MyDrive/tcc/BERT-LS-master/bert/bert-base-portuguese-uncased-vocab.txt"

  # Step 1: Save a model, configuration and vocabulary that you have fine-tuned

  # If we have a distributed model, save only the encapsulated model
  # (it was wrapped in PyTorch DistributedDataParallel or DataParallel)
  model_to_save = model.module if hasattr(model, 'module') else model

  torch.save(model_to_save.state_dict(), output_model_file)
  model_to_save.config.to_json_file(output_config_file)
  tokenizer.save_vocabulary(output_vocab_dir)

  %%shell
  mv /content/drive/MyDrive/tcc/BERT-LS-master/bert/vocab.txt /content/drive/MyDrive/tcc/BERT-LS-master/bert/bert-base-portuguese-uncased-vocab.txt
```

```python
def gerarSimplex():
  import pandas as pd
  arquivo = '/content/drive/MyDrive/tcc/BERT-LS-master/datasets/Simplex 3.0.xlsx'
  data = pd.pandas.read_excel(arquivo)

  df = data[["sentença", "palavra_dificil", "sinônimos_anotados"]]

  dict_resultado = []

  for row in df.values:
    resultado = []
    resultado.append(row[0].replace("*","").replace("'","\'").replace("´","\´"))
    resultado.append(row[1])

    sinonimos = row[2].replace('[', '').replace(']', '').replace("'", '')
    list_sinonimos = sinonimos.split(',')
    for sinonimo in list_sinonimos:
      resultado.append(sinonimo)

    dict_resultado.append(resultado)

  df = pd.DataFrame(dict_resultado)
  df.fillna(value='', inplace=True)
  df.to_csv('/content/drive/MyDrive/tcc/BERT-LS-master/datasets/Simplex30.csv', '\t', header=None, index=False, encoding='UTF-8')
```

```python
def gerarSimplex2():
  import pandas as pd
  arquivo = '/content/drive/MyDrive/tcc/BERT-LS-master/datasets/Simplex 3.0.xlsx'
  data = pd.pandas.read_excel(arquivo)

  df = data[["sentença", "palavra_dificil", "sinônimos_anotados"]]
  #df = df.iloc[942:]
  df = df.drop(941)
  dict_resultado = []

  for row in df.values:
    resultado = []
    resultado.append(row[0].replace("*","").replace("'","\'").replace("´","\´"))
    resultado.append(row[1])

    sinonimos = row[2].replace('[', '').replace(']', '').replace("'", '')
    list_sinonimos = sinonimos.split(',')
    for sinonimo in list_sinonimos:
      resultado.append(sinonimo)

    dict_resultado.append(resultado)

  df = pd.DataFrame(dict_resultado)
  df.fillna(value='', inplace=True)
  df.to_csv('/content/drive/MyDrive/tcc/BERT-LS-master/datasets/Simplex30_2.csv', '\t', header=None, index=False, encoding='UTF-8')
```

## Gerar Dataset para Treinamento do Modelo seq to seq


```python
def geraArquivosTreinamento(df, nome_arquivo):
  dict_resultado = []

  for row in df.values:
    resultado = []
    sentenca = row[0].replace("*","").replace("'","\'").replace("´","\´")
    palavra_complexa = row[1]

    nltk_sent = nltk.word_tokenize(sentenca, language='portuguese')
    for token in nltk_sent:
      resultado_linha = []
      resultado_linha.append(token)
      #resultado_linha.append(sentenca)
      #resultado_linha.append(palavra_complexa)

      if token == palavra_complexa:
        resultado_linha.append('C')
      else:
        resultado_linha.append('N')

      dict_resultado.append(resultado_linha)

    dict_resultado.append('')


  df_final = pd.DataFrame(dict_resultado)
  df_final.to_csv('/content/drive/MyDrive/tcc/BERT-LS-master/datasets/' + nome_arquivo, '\t', header=None, index=False, encoding='UTF-8')
```

```python
import pandas as pd
import nltk

arquivo = '/content/drive/MyDrive/tcc/BERT-LS-master/datasets/Simplex 3.0.xlsx'
data = pd.pandas.read_excel(arquivo)

df = data[["sentença", "palavra_dificil"]]

from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_test, df_valid = train_test_split(df_test, test_size=0.1, random_state=42)

geraArquivosTreinamento(df_train, 'simplex.train.original.tsv')
geraArquivosTreinamento(df_test,  'simplex.test.original.tsv')
geraArquivosTreinamento(df_valid, 'simplex.dev.original.tsv')
```
