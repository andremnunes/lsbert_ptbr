```python
import matplotlib.pyplot as plt  # Biblioteca para visualização de informações
import numpy as np  # Biblioteca para operações com arrays multidimensionais
import pandas as pd  # Biblioteca para análise de dados
```

```python
def graficoComparacaoJuntos(labels, dadosPotential, dadosPrecisao, dadosRecall, dadosF_Score):
  x = range(0, len(labels))

  plt.figure(figsize=(15, 7))
  plt.xticks(x, labels)
  plt.plot(dadosPotential, 'x--g', dadosPrecisao, 'o--r', dadosRecall, '*--b', dadosF_Score, '^--m', markersize=10)
  plt.legend(['Potencial', 'Precisão', 'Recall', 'F1'], loc='upper right', prop={'size':18}, bbox_to_anchor=(1.2, 0.6))
  plt.xlabel('K')
  plt.ylabel('Métricas')
  plt.rcParams['axes.titlesize'] = 24
  plt.rcParams['axes.labelsize'] = 20
  plt.rcParams['xtick.labelsize'] = 18
  plt.rcParams['ytick.labelsize'] = 18
  plt.title('Potencial, Precisão, Recall,F1 vs K')
  plt.tight_layout()
  plt.show()


def graficoComparacao(labels, data, nomeMedida, fmt):
  x = range(0, len(labels))
  y = data

  plt.figure(figsize=(35, 10))
  plt.xticks(x, labels)
  plt.plot(x, y, fmt, markersize=10)
  plt.title(nomeMedida + ' vs Modelo')
  plt.xlabel('Modelo')
  plt.ylabel(nomeMedida)
  plt.rcParams['axes.titlesize'] = 24
  plt.rcParams['axes.labelsize'] = 20
  plt.rcParams['xtick.labelsize'] = 18
  plt.rcParams['ytick.labelsize'] = 18
  plt.show()
```
